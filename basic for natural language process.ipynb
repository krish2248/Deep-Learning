# Levels of NLP

# 1. Lexical Analysis

Deals with words

Converts text into tokens

Example:

* "I love NLP"
* ‚Üí ["I", "love", "NLP"]

# 2. Syntactic Analysis (Grammar)

Checks sentence structure

Uses grammar rules

Example:

* Correct: I eat apples
* Wrong: Eat I apples

# 3. Semantic Analysis (Meaning)

Understands actual meaning

Example:

Bank ‚Üí river bank or money bank

# 4. Discourse Analysis

Understands meaning across multiple sentences

Example:

Ravi went home. He was tired.
* ‚Üí He = Ravi

# 5. Pragmatic Analysis

Understands meaning using real-world context

Example:

"Can you pass the salt?"
‚Üí Request, not ability question
